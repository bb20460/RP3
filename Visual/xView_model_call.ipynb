```python
import cv2

# Using the CV2 package to scale the 4 byte imagery to 255 pixel categories with the cv2 jet colourmap
im_loc = cv2.imread("/content/xview2_1st_place_solution/submission/zoomed_24TH_localization_disaster_prediction.png")
im_cls = cv2.imread("/content/xview2_1st_place_solution/submission/zoomed_24TH_damage_disaster_prediction.png")

img_loc_col = cv2.applyColorMap(im_loc*int(255/4), cv2.COLORMAP_JET)
img_cls_col = cv2.applyColorMap(im_cls*int(255/4), cv2.COLORMAP_JET)

# Saving the altered images to a seperate file
cv2.imwrite("/content/xview2_1st_place_solution/submission/altered/azoomed_24TH_localization_disaster_prediction.png", img_loc_col)
cv2.imwrite("/content/xview2_1st_place_solution/submission/altered/azoomed_24TH_damage_disaster_prediction.png", img_cls_col)

```
